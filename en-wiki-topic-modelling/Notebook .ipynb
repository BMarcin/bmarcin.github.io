# Wczytanie wygenerowanego wcześniej modelu

```python
from gensim.models.ldamulticore import LdaMulticore
import gensim

import logging
import sys

logging.basicConfig(format='%(levelname)s : %(message)s', level=logging.INFO)
logging.root.level = logging.INFO  # ipython sometimes messes up the logging setup; restore



id2word = gensim.corpora.Dictionary.load_from_text('data/results_wordids.txt')
mm = gensim.corpora.MmCorpus('data/results_tfidf.mm')

print(mm)
```

# LDA
Wyliczenie modelu LDA funkcją Multicore. Stosujemy workers w liczbie 3, gdyż wg. wielu opinii zwiększanie tej liczby nie przynosi większych efektów, a CPU jest obciążane na maksimum i tak.

```python
#slabe wydajnosciowo 
#czas1 = %time lda = gensim.models.ldamodel.LdaModel(corpus=mm, id2word=id2word, num_topics=200, update_every=1, passes=20)


czas1 = %time lda = LdaMulticore(mm, id2word=id2word, num_topics=200, passes=20, chunksize=70000, workers=3)
```

# Wyniki z LDA

```python
lda.print_topics(200)
```

```python
lda.save("data/lda_results")
```

# LSI


```python
czas2 = %time lsi = gensim.models.lsimodel.LsiModel(corpus=mm, id2word=id2word, num_topics=200, chunksize=70000)
```

# Wyni LSI

```python
lsi.print_topics(200)
```

```python
lsi.save("data/lsi_results")
```
